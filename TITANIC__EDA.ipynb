```python
# Cell 1 - Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cell 2 - Load Titanic Dataset
df = pd.read_csv(r"C:\Users\HP 15\Downloads\archive (2)\titanic_data.csv")

df.head()
```

```python
# Cell 3 - Dataset Shape
print("Rows and Columns:", df.shape)
```

```python
# Cell 7 - Statistical Summary
print(df.describe())
```

```python
# Cell 8 - First 10 Rows
print(df.head(10))
```

```python
# Cell 9 - Survival Count
print(df['Survived'].value_counts())
```

```python
# Cell 10 - Survival Chart
df['Survived'].value_counts().plot(kind='bar')

plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")

plt.show()
```

```python
# Cell 11 - Gender Count
print(df['Sex'].value_counts())
```

```python
# Cell 12 - Gender Chart
df['Sex'].value_counts().plot(kind='bar')

plt.title("Male vs Female")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
# Cell 13 - Passenger Class Count
print(df['Pclass'].value_counts())
```

```python
# Cell 14 - Passenger Class Chart
df['Pclass'].value_counts().plot(kind='bar')

plt.title("Passenger Class Count")
plt.xlabel("Class")
plt.ylabel("Count")

plt.show()
```

```python
# Cell 15 - Age Distribution
df['Age'].plot(kind='hist', bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")

plt.show()
```

```python
# Cell 16 - Fare Distribution
df['Fare'].plot(kind='hist', bins=20)

plt.title("Fare Distribution")
plt.xlabel("Fare")

plt.show()
```
